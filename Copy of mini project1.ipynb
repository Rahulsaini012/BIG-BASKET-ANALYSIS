# ***6. Measuring discount on a certain item***

```python
data['discount'] = ((data['market_price'] - data['sale_price']) / data['market_price']) * 100
data.head()
```

# ***7. Missing Values from the Dataset.***

```python
data.isnull().sum()
```

# ***8.  dataset according to the columns and fill them with the mean***

column-[ product ]

```python
data['product'].mode()[0]
data['product']=data['product'].fillna(data['product'].mode()[0])
```

column-[ brand ]

```python
data['brand'].mode()[0]
data['brand']=data['brand'].fillna(data['brand'].mode()[0])
```

column-[ sale_price ]

```python
data['sale_price'].mean()
data['sale_price']=data['sale_price'].fillna(data['sale_price'].mean())
```

column-[ rating ]

```python
data['rating'].mean()
data['rating']=data['rating'].fillna(data['rating'].mean())
```

column-[ description	]

```python
data['description'].mode()[0]
data['description']=data['description'].fillna(data['description'].mode()[0])
data.isna().sum()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Set up the Seaborn style for better visuals
sb.set(style="whitegrid")

# 1. Distribution of Sale Prices
plt.figure(figsize=(10, 6))
sb.histplot(data['sale_price'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Sale Prices")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Distribution of Market Prices
plt.figure(figsize=(10, 6))
sb.histplot(data['market_price'], bins=30, kde=True, color='green')
plt.title("Distribution of Market Prices")
plt.xlabel("Market Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Boxplot to show the spread and outliers of sale prices
plt.figure(figsize=(8, 6))
sb.boxplot(x=data['sale_price'], color='orange')
plt.title("Sale Price Boxplot")
plt.xlabel("Sale Price")
plt.show()
```

```python
# 4. Boxplot to show the spread and outliers of market prices
plt.figure(figsize=(8, 6))
sb.boxplot(x=data['market_price'], color='purple')
plt.title("Market Price Boxplot")
plt.xlabel("Market Price")
plt.show()
```

```python

# 5. Countplot for Product Categories
plt.figure(figsize=(12, 6))
sb.countplot(x='category', data=data, palette='Set2')
plt.title("Product Categories Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Countplot for Product Types
plt.figure(figsize=(10, 6))
sb.countplot(x='type', data=data, palette='muted')
plt.title("Product Types Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Scatterplot between Market Price and Sale Price
plt.figure(figsize=(10, 6))
sb.scatterplot(x='market_price', y='sale_price', data=data, color='red')
plt.title("Market Price vs Sale Price")
plt.xlabel("Market Price")
plt.ylabel("Sale Price")
plt.show()
```

```python
# 8. Rating Distribution
plt.figure(figsize=(10, 6))
sb.histplot(data['rating'], bins=30, kde=True, color='purple')
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# 9. Countplot for Brands (Top 10)
plt.figure(figsize=(12, 6))
top_brands = data['brand'].value_counts().head(10).index
sb.countplot(x='brand', data=data[data['brand'].isin(top_brands)], order=top_brands, palette='coolwarm')
plt.title("Top 10 Brands by Count")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
